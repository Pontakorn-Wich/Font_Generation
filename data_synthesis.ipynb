# Data Synthesis for Font Characters

This notebook downloads open fonts from Google Fonts, renders character images, and exports:
- `data/labels.csv` with `label_character, font_name, picture_path`
- `data/char_mapping.csv` with same-character source/target pairs across fonts

```python
from pathlib import Path
from data_synthesis import download_google_fonts, render_character_images
```

```python
ROOT = Path('.')
FONT_DIR = ROOT / 'data' / 'fonts'
IMAGE_DIR = ROOT / 'data' / 'images'
LABELS_CSV = ROOT / 'data' / 'labels.csv'
MAPPING_CSV = ROOT / 'data' / 'char_mapping.csv'
CHARS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789'
IMAGE_SIZE = 128
FONT_SIZE = 96
```

```python
downloaded = download_google_fonts(FONT_DIR)
print(f'Downloaded {downloaded} new font files to {FONT_DIR}')
```

```python
stats = render_character_images(
    font_dir=FONT_DIR,
    image_dir=IMAGE_DIR,
    labels_csv_path=LABELS_CSV,
    mapping_csv_path=MAPPING_CSV,
    chars=CHARS,
    image_size=IMAGE_SIZE,
    font_size=FONT_SIZE,
)
stats
```

```python
print('Main CSV:', LABELS_CSV)
print('Mapping CSV:', MAPPING_CSV)
print('Rows in labels.csv:', stats['labels_rows'])
print('Rows in char_mapping.csv:', stats['mapping_rows'])
```
